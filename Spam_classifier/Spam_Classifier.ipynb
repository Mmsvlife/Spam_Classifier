```python
# reading the data from the table using pandas 

import pandas as pd

df = pd.read_table("smsspamcollection/SMSSpamCollection", 
                   sep = '\t', header = None, names=['label', 'sms_message'])
df.head()
```

```python
# converting the spam:1 and Ham:0

df['label'] = df.label.map({'ham':0, 'spam':1})
print(df.shape)
df.head() # returns (rows, columns)
```

```python
# converting all the documents to lowercase.
documents = ['Hello! How are you?',
            'Wassup! bro how you doing?',
             "let's hangout.",
            "let's count the numbers 123go."]

lower_case_documents = []
for i in documents:
    lower_case_documents.append(i.lower())

lower_case_documents
```

```python
# converting the original document
lower_document = []
for i in df['sms_message']:
    i = i.lower()
    lower_document.append(i)
#     print(i)
lower_document
# df["sms_message"] = lower_document
# df.head()
```

```python
# removing punctuation from the few document.
sans_punctuation_document= []
import string

for i in lower_case_documents:
    sans_punctuation_document.append(i.translate(str.maketrans('','',string.punctuation)))
sans_punctuation_document
```

```python
#removing the Punctuation from the original document
no_punctuation = []

for i in lower_document:
    no_punctuation.append(i.translate(str.maketrans('','', string.punctuation)))

print(type(no_punctuation))
```

```python
#removing the numbers from the document using string
from string import digits
no_digits = []
remove_digits = str.maketrans("","",digits)
for i in sans_punctuation_document:
    no_digits.append(i.translate(remove_digits))
no_digits
```

```python
# removing the numbers from the original messages
final_messages = []

for i in no_punctuation:
    final_messages.append(i.translate(str.maketrans("","",digits)))

final_messages
```

```python
df["sms_message"] = final_messages
df.head()
```

```python
#tokenizing the sentences
preprocess_document = []
for i in no_digits:
    preprocess_document.append(i.split(" "))
preprocess_document
```

```python
#tokenizing or spliting the original document
token_document = []
for i in final_messages:
    token_document.append(i.split(" "))
token_document
```

```python
frequency_list = []
import pprint
from collections import Counter
for i in token_document:
    frequency_counts = Counter(i)
    frequency_list.append(frequency_counts)

frequency_list
```

```python
from sklearn.feature_extraction.text import CountVectorizer
count_vector = CountVectorizer()
```

```python
count_vector.fit(lower_document)
count_vector.get_feature_names()
```

```python
from sklearn.cross_validation import train_test_split

x_train,x_test,y_train,y_test = train_test_split(df['sms_message'],df['label'], random_state = 1)
print("Number of rows in total set: {}".format(df.shape[0]))
print("Number of train set: {}".format(x_train.shape[0]))
print("Number of test set: {}".format(x_test.shape[0]))
```

```python
count_vector = CountVectorizer()
training_data = count_vector.fit_transform(x_train)
testing_data = count_vector.transform(x_test)
```

```python
from sklearn.naive_bayes import MultinomialNB
naive_bayes = MultinomialNB()
naive_bayes.fit(training_data,y_train)
```

```python
# predictions = naive_bayes.predict(testing_data)
predictions = naive_bayes.predict(testing_data)
```

```python
from sklearn.metrics import accuracy_score, precision_score,recall_score, f1_score

print("Accuracy Score:{}".format(accuracy_score(y_test,predictions)))
print('Precision score: ', format(precision_score(y_test, predictions)))
print('Recall score: ', format(recall_score(y_test, predictions)))
print('F1 score: ', format(f1_score(y_test, predictions)))
```
